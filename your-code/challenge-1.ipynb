#### From the Fitbit data, we want to visually understand:

How does the average number of steps change by month? Use the appropriate visualization to show the median steps by month. Is Fitbitter more active on weekend or workdays?
**Hints**:

* Use Months_encoded and Week_or_Weekend columns.
* Set your size figure to 12,4
* Explore plt.xticks
* Save your figures in a folder called `figures` in your repo. 

```python
plt.figure(figsize=(12,4))
sns.barplot(x="Months_encoded", y="Steps",data=fitbit,hue="Work_or_Weekend")
plt.xlabel("Months")
plt.title("Average steps monthly")
plt.ylabel('Num of Steps')
plt.savefig("avg_steps_monthly.png")
plt.show()

```

#### Write a loop to plot 3 scatter plots of the following features:

* Minutes Lightly Active vs Steps    
* Minutes Very Active vs Steps    
* Minutes Sedentary vs Steps  

```python
plt.figure(figsize=(12,4))

fields_list=["Minutes Lightly Active", "Minutes Very Active","Minutes Sedentary"]
for i in fields_list:
    sns.scatterplot(x=fitbit[i], y=fitbit["Steps"]);
```

```python

fields_list=["Minutes Lightly Active", "Minutes Very Active","Minutes Sedentary"]
for i in fields_list:
    plt.figure(figsize=(12,4))

    sns.scatterplot(x=fitbit[i], y=fitbit["Steps"]);
    plt.show()

```
